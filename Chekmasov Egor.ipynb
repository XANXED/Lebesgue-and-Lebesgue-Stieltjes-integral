# Численный метод

### Импорт библиотек и создание функций

```python
import numpy as np
import matplotlib.pyplot as plt

# Зададим исходную функцию, с которой будем сравнивать приближение
def f(x):
    return x**2

# Сделаем простую функцию f_n: на каждом отрезке берем левый конец
def f_n(x, n):
    # Разобьем [0; 4] на 2^n одинаковых интервалов
    num_intervals = 2**n

    # Отдельно обработаем число, потому что для него не работают векторные операции NumPy
    if np.isscalar(x):
        k = int(x * num_intervals) // 4
        # Ограничим индекс, чтобы не выйти за последний интервал
        k = max(0, min(k, num_intervals - 1))
        x_k_min_1 = 4 * k / num_intervals
    else:
        # Для массива сразу найдем шаг и номера интервалов
        dx = 4 / num_intervals
        k = np.clip(np.floor(x / dx).astype(int), 0, num_intervals - 1)
        x_k_min_1 = k * dx

    # Вернем значение f в левом конце интервала
    return x_k_min_1**2
# Запомним точное значение интеграла, чтобы потом сравнить
integral_real_1 = 64/3
integral_real_2 = 5
```

## 2.1 Изобразим график f и f_n (для сравнения) при n от 1 до 10

```python
# Подключим инструменты для анимации и показа в ноутбуке
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Настроим область графика перед первым кадром
def init():
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 16)
    return step_plot,

# Для каждого n будут пересчитываться значения ступенчатой функции
def update(n):
    count = len(x_vals)
    y_step = [0] * count

    for i in range(count):
        x = x_vals[i]
        # В каждой точке возьмем значение f_n
        y_step[i] = f_n(x, n)
    step_plot.set_data(x_vals, y_step)
    ax.set_title(f'f и f_n при $n = {n}$ (количество интеравлов = {2**n})')
    return step_plot,

# Подготовим окно и сетку точек на отрезок [0; 4]
fig, ax = plt.subplots(figsize=(10, 6))
x_vals = np.linspace(0, 4, 1000)

# Сначала нарисуем исходную f(x), потом добавим пустую линию для f_n
ax.plot(x_vals, f(x_vals), color='black', alpha=0.3, label='$f(x) = x^2$')
step_plot, = ax.step([], [], where='post', color='red', label='Простая функция $f_n$')

# Соберем анимацию для n от 1 до 10
ani = FuncAnimation(fig, update, frames=range(1, 11), init_func=init, blit=True)

# Закроем статичную картинку и выведем именно анимацию
plt.close()
HTML(ani.to_jshtml())
```

## 2.2
## Вычислим интеграл Лебега для $f_n$ при разных n и сравним с результатом, полученным в пункте 1.3

```python
# Посчитаем интеграл Лебега для f_n по готовой формуле
def integral_l(n):
    intervals = 2**n
    return 64 / 3 - (32 / intervals) + (32 / (3 * intervals**2))

# Прогоним несколько n и сравним с точным интегралом
for n in [10, 100, 1000]:
    res = integral_l(n)
    # Посчитаем модуль отклонения от аналитического значения
    difference = abs(res - integral_real_1)
    print(f"n = {n:4}: Интеграл = {res:.8f} (Отклонение от аналитического результата: {difference:.8e})")
```

## 2.3
## Вычислим интеграл Лебега-Стилтьеса для $f_n$ при разных n и сравним с результатом, полученным в пункте 1.5

```python
# Посчитаем интеграл Лебега-Стилтьеса через точки скачков
def integral_ls(n):
    # Зададим точки скачков и их веса
    jump_points = [1, 2]
    jump_weights = [1, 1]
    res = 0
    # Пройдемся по всем скачкам и суммируем вклад каждой точки
    for i in range(len(jump_points)):
        x = jump_points[i]
        w = jump_weights[i]
        # Берем значение функции в точке скачка
        value = f_n(x, n)

        res = res + value * w
    return res

# Проверим результат для больших n и сравним с точным значением
for n in [10000, 100000, 1000000]:
    res = integral_ls(n)
    # Посчитаем отклонение от аналитического результата
    difference = abs(res - integral_real_2)
    print(f"n = {n:4}: Интеграл = {res:.8f} (Отклонение от аналитического результата: {difference:.8e})")
```
